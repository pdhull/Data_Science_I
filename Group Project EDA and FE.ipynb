Outlier removal

```python
# List of numerical columns

num_cols = ['Age', 'Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
            'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date',
            'Num_of_Delayed_Payment', 'Changed_Credit_Limit', 'Num_Credit_Inquiries',
            'Outstanding_Debt', 'Credit_Utilization_Ratio', 'Total_Age_in_Months',
            'Total_EMI_per_month', 'Amount_invested_monthly',
            'Monthly_Balance']
```

```python
# Identify and remove outliers using IQR method

for col in num_cols:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1

    # Define limits

    lower_limit = Q1 - 2.0 * IQR
    upper_limit = Q3 + 2.0 * IQR

    # Identify outliers

    outliers = df_cleaned[(df_cleaned[col] < lower_limit) | (df_cleaned[col] > upper_limit)]

    # Remove outliers

    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_limit) & (df_cleaned[col] <= upper_limit)]
```

Feature Engineering

```python
# Create interaction feature

df_cleaned['Income_Salary_Interaction'] = df_cleaned['Annual_Income'] * df_cleaned['Monthly_Inhand_Salary']
```

```python
# Create polynomial features

df_cleaned['Num_Bank_Accounts_Squared'] = df_cleaned['Num_Bank_Accounts']**2
df_cleaned['Num_Credit_Card_Squared'] = df_cleaned['Num_Credit_Card']**2
```

```python
# Bin Age into categories

bins = [18, 35, 55, df_cleaned['Age'].max()]
labels = ['Young', 'Middle-aged', 'Old']
df_cleaned['Age_Category'] = pd.cut(df_cleaned['Age'], bins=bins, labels=labels)

```
