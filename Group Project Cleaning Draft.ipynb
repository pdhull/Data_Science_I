```python
df.drop(df[df['Age'] == '-500'].index, inplace = True)
```

```python
df['Age'] = df['Age'].str.replace('-','')
df['Age'] = df['Age'].str.replace('_','')
```

```python
# Deleting (-) and (_) in Column
df['Annual_Income'] = df['Annual_Income'].str.replace('-','')
df['Annual_Income'] = df['Annual_Income'].str.replace('_','')

# Deleting (-) and (_) in Columndata1['Num_of_Loan'] = data1['Num_of_Loan'].str.replace('-','')
df['Num_of_Loan'] = df['Num_of_Loan'].str.replace('_','')

# Deleting (-) and (_) in Column
df['Changed_Credit_Limit'] = df['Changed_Credit_Limit'].str.replace('-','0')
df['Changed_Credit_Limit'] = df['Changed_Credit_Limit'].str.replace('_','0')

# Deleting (-) and (_) in Column
df['Outstanding_Debt'] = df['Outstanding_Debt'].str.replace('-','')
df['Outstanding_Debt'] = df['Outstanding_Debt'].str.replace('_','')

# Deleting (-) and (_) in Column
df['Monthly_Balance'] = df['Monthly_Balance'].str.replace('-','')
df['Monthly_Balance'] = df['Monthly_Balance'].str.replace('_','')

# Replacing (!@9#%8) to (Unknown) in Column
df['Payment_Behaviour'] = df['Payment_Behaviour'].str.replace('!@9#%8','Unknown')
```

```python
df['Age'] = df['Age'].astype(int)
df['Annual_Income'] = df['Annual_Income'].astype(float)
df['Num_of_Loan'] = df['Num_of_Loan'].astype(int)
df['Changed_Credit_Limit'] = df['Changed_Credit_Limit'].astype(float)
df['Outstanding_Debt'] = df['Outstanding_Debt'].astype(float)
df['Monthly_Balance'] = df['Monthly_Balance'].astype(float)
```

```python
df.info()
```

```python
# Drop Unimportant Columns
df = df.drop(columns=['ID','Month','Name','SSN',
                           'Type_of_Loan','Num_of_Delayed_Payment','Credit_History_Age','Amount_invested_monthly',
                        'Payment_Behaviour'])
```

```python
df.info()
```

```python
# Replacing outlier of Age with Ffill
df['Age'].where((df['Age']<=80),np.nan, inplace=True)
df['Age'] = df['Age'].fillna(method='ffill')
df
```

```python
# Replacing outlier of Age with Ffill
df['Age'].where((df['Age']>=18),np.nan, inplace=True)
df['Age'] = df['Age'].fillna(method='ffill')
df
```

```python
# Handling missing values with mean
df['Monthly_Balance'] = df['Monthly_Balance'].fillna(df['Monthly_Balance'].mean())
df['Num_Credit_Inquiries'] = df['Num_Credit_Inquiries'].fillna(df['Num_Credit_Inquiries'].mean())

# Handling missing values with ffill
df['Occupation'] = df['Occupation'].replace('_______', method='ffill')
df.head()
```

```python
# Handling missing values with bfill
df['Credit_Mix'] = df['Credit_Mix'].replace('_', method='bfill')
df.head()
```

```python
df.groupby(['Credit_Mix']).count()[['Occupation']]
```

```python
from sklearn.preprocessing import LabelEncoder
lE = LabelEncoder()
```

```python
df['Credit_Mix'] = df['Credit_Mix'].str.replace('Bad','0')
df['Credit_Mix'] = df['Credit_Mix'].str.replace('Standard','1')
df['Credit_Mix'] = df['Credit_Mix'].str.replace('Good','2')
df['Credit_Mix'] = df['Credit_Mix'].astype(int)
df.groupby(['Credit_Mix']).count()[['Occupation']]
```

```python
df['Payment_of_Min_Amount'] = lE.fit_transform(df['Payment_of_Min_Amount'])
df.groupby(['Payment_of_Min_Amount']).count()[['Occupation']]
```

```python
df.groupby(['Credit_Score']).count()[['Occupation']]
```

```python
df['Credit_Score'] = df['Credit_Score'].str.replace('Poor','0')
df['Credit_Score'] = df['Credit_Score'].str.replace('Standard','1')
df['Credit_Score'] = df['Credit_Score'].str.replace('Good','2')
df['Credit_Score'] = df['Credit_Score'].astype(int)
df.groupby(['Credit_Score']).count()[['Occupation']]
```

```python
df = pd.get_dummies(df, columns=['Occupation'], prefix=['Occ_'])
df.head()
```
